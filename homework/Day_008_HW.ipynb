```python
import os
import numpy as np
import pandas as pd


#設定資料讀取路徑

dir_data = 'desktop/data/'
```

```python
f_app_train = os.path.join(dir_data,'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
import matplotlib.pyplot as mp
#先了解資料型態
print(app_train.shape)
#先了解資料欄位
app_train.head()
```

```python
#AMT_INCOME_TOTAL  平均數
app_train_AMT_INCOME_TOTAL = app_train['AMT_INCOME_TOTAL']
#轉為數組
df_app_train_AMT_INCOME_TOTAL = pd.DataFrame(app_train_AMT_INCOME_TOTAL)
#print(df_app_train_AMT_INCOME_TOTAL)

print('app_train_AMT_INCOME_TOTAL的平均數為',app_train_AMT_INCOME_TOTAL.mean())

#MT_INCOME_TOTAL  平方差

app_train_AMT_INCOME_TOTAL_VAR = np.var(app_train_AMT_INCOME_TOTAL)
print('app_train_AMT_INCOME_TOTAL_VAR的平方差為',app_train_AMT_INCOME_TOTAL_VAR)

#AMT_INCOME_TOTAL  標準差
app_train_AMT_INCOME_TOTAL_STD = np.std(app_train_AMT_INCOME_TOTAL,ddof=1)
print('app_train_AMT_INCOME_TOTAL_STD的標準差為',app_train_AMT_INCOME_TOTAL_STD)

df_app_train_AMT_INCOME_TOTAL_STD = np.std(df_app_train_AMT_INCOME_TOTAL,ddof=1)
print('df_app_train_AMT_INCOME_TOTAL_STD的標準差為',df_app_train_AMT_INCOME_TOTAL_STD)
```

```python
#畫出任意欄位直方圖（app_train_AMT_INCOME_TOTAL）

mp.hist(app_train['AMT_INCOME_TOTAL'],bins=[5000,10000,30000,50000,75000,100000,150000,200000,250000,300000])
mp.show()



```
